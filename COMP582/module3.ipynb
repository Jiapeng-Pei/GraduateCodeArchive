### 4. Insertion sort is better than selection sort.
  In this case, the time complexity of selection sort is still $O(N^2)$, since the inner loop doesn't terminate until it reaches the end of the array.

  However, the time complexity of insertion sort is $O(N)$. The outer loop still traverses $N$ elements. But inside the outer loop it takes $O(1)$ since the subarray is already sorted. $O(N)$ is better than $O(N^2)$.


### 5. Both algorithms perform the same.
  The time complexity of selection sort is still $O(N^2)$. 

  For the insertion sort, it takes $1+2+3+...+(N-1)$ compare and swap operations. The time complexity of compare and swap is $O(1)$, so the overall complexity is $O(N^2)$. So they perform the same.

### 8. This h-sequence results in worst case time complexity.
  If we only use even h-sequence numbers before 1, like: 2, 4, 8, 16..., then we only sort numbers on even indexs and odd indexes respectively. And we don not know the the size relationship between even and odd elements. This may results in bad complexity when performing shell sort, which is $O(N^2)$. Take the following array as as example:

  1 6 2 7 3 8 4 9 5 10
   
  If we use h-sequence 1, 2, 4 to sort this array, then before using 1, this array doesn't need to be sorted. But in the last step, elements on even indexes needs to move $i/2$ times, where $i$ denotes their index. 
  
  $T(1) = 0 + 1 + 2 + ... + n/2-1$, which equals to $O(N^2)$

### 9. Because insertion sort has better complexity compared to selection sort.
  - Here's how we perform shell sort:
    By specifying the step size, the original array is divided into step independent sub-arrays, and then these sub-arrays are sorted separately by insertion sort (i.e. group sort), which we call a step ordered array. When the step is very large, we can move the elements to a very far position at once to create a convenience for the next smaller step order; keep reducing the step size, repeat the above process to build a step ordered array to achieve the purpose of local order. When the step is finally 1 to do the last step order, is the insertion sort we are usually familiar with, because the array has been grouped many times by the larger step sorted, at this time only need fewer times the elements move to achieve the whole array global order. 
  
  - From the above analysis, we can find that in each step we are sorting a partially sorted subarray (almost in order), where the efficiency of insertion sort is much better than that of selection sort.





````python
def removeDuplicates(nums):
    nums.sort()
    ret = []
    i = 0
    while (i < len(nums)):
        ret.append(nums[i])
        i += 1
        while (i < len(nums) and nums[i] == nums[i-1]): i += 1
        
    return ret;

print(removeDuplicates([1, 9, 9, 9, 8, 7, 7, 2, 3, 5, 2, 6, 1, 4]))
````

````python
def findJumbles(words):
    pairs = []
    for word in words:
        chars = ''.join(sorted(word))
        pairs.append((chars, word))
        
    pairs = sorted(pairs, key=lambda x: x[0])
    i = 0
    ret = []
    while (i < len(pairs)):
        ret.append(pairs[i][1])
        i += 1
        while (i < len(pairs) and pairs[i][0] == pairs[i-1][0]): 
            ret.append(pairs[i][1])
            i += 1
        ret.append('\n')
        
    return ret

ret = findJumbles(['racing', 'secura', 'saucer', 'caring', 'random'])
for str in ret : 
    if str != '\n': print(str, end=' ')
    else: print()
````
